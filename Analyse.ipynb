```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def top_cv_volatilite_graph(df, top_n=5):
    # Préparation des données
    
    df_6mois = df[df.index.month <= 6]
    df_6mois = df_6mois.apply(pd.to_numeric, errors='coerce')

    # Coefficient de variation (CV)
    mean_prices = df_6mois.mean()
    std_prices = df_6mois.std()
    cv = std_prices / mean_prices

    # Volatilité (écart-type des rendements log)
    log_returns = np.log(df_6mois / df_6mois.shift(1))
    volatilite = log_returns.std()

    # Top actions
    top_cv = cv.sort_values(ascending=False).head(top_n)
    top_vol = volatilite.sort_values(ascending=False).head(top_n)

    # Création graphique avec deux sous-graphes
    fig = make_subplots(rows=1, cols=2, subplot_titles=("Top Coefficient de Variation", "Top Volatilité"))

    # Bar chart CV
    fig.add_trace(go.Bar(
        x=top_cv.values[::-1],
        y=top_cv.index[::-1],
        orientation='h',
        name='Coefficient de Variation',
        marker_color='indianred',
        text=[f"{v:.2f}" for v in top_cv.values[::-1]],
        textposition='auto'
    ), row=1, col=1)

    # Bar chart Volatilité
    fig.add_trace(go.Bar(
        x=top_vol.values[::-1],
        y=top_vol.index[::-1],
        orientation='h',
        name='Volatilité (log-retours)',
        marker_color='lightslategray',
        text=[f"{v:.4f}" for v in top_vol.values[::-1]],
        textposition='auto'
    ), row=1, col=2)

    fig.update_layout(
        height=500,
        width=900,
        title_text="Top actions sur les 6 premiers mois : Coefficient de Variation vs Volatilité",
        showlegend=False
    )

    fig.show()

# Exemple d’utilisation avec ton DataFrame `df`
# top_cv_volatilite_graph(df, top_n=5)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import linprog

def portefeuille_diversifie(df, capital, annee, cible_rendement=0.10, poids_max=0.3):


    # Filtrer sur l’année
    df_annee = df[df.index.year == annee]
    df_annee = df_annee.ffill().bfill()

    if df_annee.empty or df_annee.shape[0] < 2:
        print(f"Aucune donnée suffisante pour l'année {annee}")
        return

    # Calcul des rendements
    rendements = (df_annee.iloc[-1] / df_annee.iloc[0]) - 1
    actions_eligibles = rendements[rendements >= cible_rendement]

    if actions_eligibles.empty:
        print(f"Aucune action n'a atteint un rendement ≥ {cible_rendement:.0%} en {annee}.")
        return

    noms = actions_eligibles.index.tolist()
    R = actions_eligibles.values

    n = len(R)
    c = -R  # on maximise R.w

    # Contraintes
    A_eq = [np.ones(n)]  # somme des poids = 1
    b_eq = [1]

    # Rendement cible : R.w ≥ cible_rendement → -R.w ≤ -cible_rendement
    A_ub = [-R]
    b_ub = [-cible_rendement]

    bounds = [(0, poids_max) for _ in range(n)]

    res = linprog(c, A_eq=A_eq, b_eq=b_eq, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

    if res.success:
        poids = res.x
        allocation = capital * poids
        rendement_portefeuille = np.dot(R, poids)

        print(f"✅ Rendement attendu du portefeuille en {annee} : {rendement_portefeuille:.2%}")
        print(f"📊 Diversification garantie (max {poids_max*100:.0f}% par ligne)")
        print("💼 Allocation optimale :")
        for i, a in enumerate(noms):
            print(f"- {a} : {allocation[i]:,.0f} FCFA ({poids[i]*100:.1f}%)")
    else:
        print("❌ Optimisation échouée : essaye d’ajuster le seuil ou le poids max.")

# Appel : capital = 10 millions, seuil = 10%, max 30% par ligne
portefeuille_diversifie(df, capital=10_000_000, annee=2025, cible_rendement=0.10, poids_max=0.3)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
from datetime import datetime

def portefeuille(df, capital, cible_rendement=0.10, poids_max=0.25, correlation_max=0.8):
    
    # 1. Filtrer période janvier 2024 à aujourd’hui
    df = df[(df.index >= pd.Timestamp('2024-07-01')) & (df.index <= pd.Timestamp(datetime.now().date()))]
    df = df.ffill().bfill()

    if df.empty or df.shape[0] < 2:
        print("❌ Données insuffisantes.")
        return

    # 2. Rendements mensuel
    #returns = df.pct_change().dropna()
    returns = df.resample('M').last().pct_change().dropna()

    # 3. Rendement total sur la période
    cum_returns = (df.iloc[-1] / df.iloc[0]) - 1
    rentables = cum_returns[cum_returns >= cible_rendement]

    if rentables.empty:
        print("❌ Aucune action n'a un rendement ≥ 10%.")
        return

    # 4. Corrélation
    returns_filtered = returns[rentables.index]
    corr_matrix = returns_filtered.corr()
    avg_corr = corr_matrix.apply(lambda x: x.drop(x.name).mean())
    faiblement_correlees = avg_corr[avg_corr <= correlation_max].index.tolist()

    if not faiblement_correlees:
        print("❌ Aucune action avec corrélation moyenne ≤ seuil.")
        return

    # 5. Variables
    rets = returns[faiblement_correlees]
    mean_returns = rets.mean()
    cov_matrix = rets.cov()
    n = len(faiblement_correlees)

    w = cp.Variable(n)
    risk = cp.quad_form(w, cov_matrix.values)
    expected_return = mean_returns.values @ w

    # 6. Problème d’optimisation
    constraints = [
        cp.sum(w) == 1,
        expected_return >= cible_rendement / 12,  # annualisé → mensuel
        w >= 0,
        w <= poids_max
    ]
    prob = cp.Problem(cp.Minimize(risk), constraints)
    prob.solve()

    if w.value is None:
        print("❌ Optimisation échouée.")
        return

    poids = w.value
    poids = np.round(poids, 4)
    expected_return = mean_returns.values @ poids
    rendement_annuel = (1 + expected_return)**12 -1  # annualisation correcte

    portefeuille = pd.Series(poids, index=faiblement_correlees)
    portefeuille = portefeuille[portefeuille > 0]

    print(f" Rendement projeté annualisé : {rendement_annuel *100:.4f}%")
    print(f" Rendement mensuel experé: {expected_return*100 :.4f}%")
    print(f" Diversification réelle : poids max = {poids_max*100:.0f}%, corrélation max = {correlation_max}")
    print("💼 Allocation optimale :")
    for act, p in portefeuille.items():
        montant = p * capital
        print(f"- {act} : {montant:,.0f} FCFA ({p*100:.2f}%)")


# Appel
portefeuille(df, capital=10_000_000, cible_rendement=0.11, poids_max=0.20, correlation_max=0.35)
```
