```python
from faker import Faker
import random
from datetime import timedelta
import pandas as pd

fake = Faker('fr_FR')

def generate_bond(index):
    taux = round(random.uniform(3.0, 9.0), 2)
    start_date = fake.date_between(start_date='-15y', end_date='today')
    years_duration = random.randint(3, 10)
    end_date = start_date.replace(year=start_date.year + years_duration)
    periodicite = "Semestriel"
    nbre_periodes = years_duration * 2
    montant_emis = 10000
    valeur_nominale = 10000

    titre = f"OBLIG {fake.company()} {taux}% {start_date.year}-{end_date.year}"
    
    return {
        "Titre": titre,
        "Periodicite": periodicite,
        "DateJouissance": start_date.strftime('%Y-%m-%d'),
        "DateFin": end_date.strftime('%Y-%m-%m'),
        "NbreDifferes": 5,
        "MontantEmis": montant_emis,
        "NbrePeriodes": nbre_periodes,
        "TauxInteretsNet": taux,
        "ValeurNominale": valeur_nominale
    }

# Générer 50 obligations
obligations = [generate_bond(i) for i in range(50)]

# Convertir en DataFrame
df = pd.DataFrame(obligations)

# Sauvegarde CSV avec séparateur virgule
df.to_csv("obligations_fictives.csv", index=False, sep=',')

# Aperçu

```

```python
import os

# Chemin du dossier contenant les fichiers
dossier = 'c:\\Users\\marcel.anee\\Desktop\\simul_portefeuil\\data_action'

# Préfixe à supprimer
prefixe = "cotation_"

# Parcours des fichiers dans le dossier
for nom_fichier in os.listdir(dossier):
    # Vérifie si le nom commence par "quotation_"
    if nom_fichier.startswith(prefixe):
        # Nouveau nom sans le préfixe
        nouveau_nom = nom_fichier[len(prefixe):]
        
        # Chemins complets
        ancien_chemin = os.path.join(dossier, nom_fichier)
        nouveau_chemin = os.path.join(dossier, nouveau_nom)
        
        # Renomme le fichier
        os.rename(ancien_chemin, nouveau_chemin)
        
        print(f"{nom_fichier} → {nouveau_nom}")
```
